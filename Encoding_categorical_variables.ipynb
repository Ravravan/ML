#COLUMN TRANSFORMER

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
```

```python
df=pd.read_csv('covid_toy.csv')
```

```python
df.head()
```

```python
df['cough'].value_counts()
```

```python
df['city'].value_counts()
```

```python
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test,y_train, _test = train_test_split(df.drop(columns=['has_covid']), df['has_covid'], test_size=0.2)
```

```python
x_train
```

```python

#without columns transformer

#adding simple imputer to fever col

si =SimpleImputer()
x_train_fever = si.fit_transform(x_train[['fever']])

#also the test dats

x_test_fever =si.fit_transform(x_test[['fever']])
x_train_fever
```

```python
x_train_fever.shape
```

```python
#ordinal encoding -> cough

oe =OrdinalEncoder(categories=[['Mild', 'Strong']])
x_train_cough = oe.fit_transform(x_train[['cough']])

#also  the tesg data

x_test_cough = oe.transform(x_test[['cough']])
x_train_cough
```

```python
x_train_cough.shape
```

```python
#onhotencoding -> gender,city
ohe=OneHotEncoder(drop='first')
x_train_gender_city = ohe.fit_transform(x_train[['gender','city']]).toarray()

#also the test data
x_test_gender_city =ohe.fit_transform(x_test[['gender','city']]).toarray()
x_train_gender_city.shape
```

```python
#extrscing age
x_train_age = x_train.drop(columns=['gender','fever','city','cough']).values
x_test_age = x_test.drop(columns=['gender','fever','city', 'cough']).values
```

```python
x_train_age
```

```python
x_train_age.shape
```

```python
x_train_transformed = np.concatenate((x_train_age, x_train_fever, x_train_gender_city, x_train_cough), axis=1)
x_test_transformed = np.concatenate((x_test_age, x_test_fever, x_test_gender_city, x_test_cough), axis =1)
```

```python
x_train_transformed.shape
```

```python
#with column transformer
from sklearn.compose import ColumnTransformer
```

```python
transformer = ColumnTransformer(transformers=[
    ('tnf1', SimpleImputer(),['fever']),
    ('tnf2', OrdinalEncoder(categories=[['Mild', 'Strong']]), ['cough']),
    ('tnf3', OneHotEncoder(sparse_output=False, drop='first'), ['gender','city']),

], remainder='passthrough')
```

```python
transformer.fit_transform(x_train)
```

```python
transformer.fit_transform(x_train).shape
```

```python
transformer.transform(x_test)
```

```python
transformer.transform(x_test).shape
```
