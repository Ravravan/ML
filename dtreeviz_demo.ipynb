```python
import graphviz.backend as be
```

```python
from sklearn.datasets import *
from dtreeviz.trees import *
from IPython.display import Image, display_svg, SVG
```

```python
from sklearn import tree
clas = tree.DecisionTreeClassifier()
iris =load_iris()

X_train = iris.data
y_train = iris.target
clas.fit(X_train,y_train)
```

```python
from sklearn.tree import plot_tree
plot_tree(clas)
```

#classification

```python
from dtreeviz import model as dtreeviz_model

m = dtreeviz_model(clas,
                   X_train,
                   y_train,
                   target_name='species',
                   feature_names=iris.feature_names,
                   class_names=iris.target_names)

viz = m.view(scale = 2)
viz
```

#Horizontal Decision Tree

```python
from dtreeviz import model as dtreeviz_model

m = dtreeviz_model(clas,
                   X_train,
                   y_train,
                   target_name='species',
                   feature_names=iris.feature_names,
                   class_names=iris.target_names)

viz = m.view(scale=1.5,
             orientation='LR')
viz
```

#node number

```python
from dtreeviz import model as dtreeviz_model

m = dtreeviz_model(clas,
                   X_train,
                   y_train,
                   target_name='species', # Corrected for Iris dataset
                   feature_names=iris.feature_names,
                   class_names=iris.target_names) # Using iris.target_names for robustness

viz = m.view(histtype='barstacked',
             scale=1.5,
             orientation='LR',
             show_node_labels=True)
viz
```
